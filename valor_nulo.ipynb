Eliminar filas inecesarias

```python
import pandas as pd

df = pd.read_csv("CDMX.csv", encoding="ISO-8859-1")

# Reemplazar valores nulos por la media de cada columna numérica
df.fillna(df.mean(numeric_only=True), inplace=True)

# Eliminar filas con más del 50% de valores nulos
df.dropna(thresh=int(0.5 * len(df.columns)), inplace=True)

# Eliminar filas duplicadas
df.drop_duplicates(inplace=True)

columnas_relevantes = df.select_dtypes(include=['number']).columns[:10]

# Mostrar las 10 filas con mas cambois
df_relevantes = df[columnas_relevantes].head(10)

print(df_relevantes)
```
